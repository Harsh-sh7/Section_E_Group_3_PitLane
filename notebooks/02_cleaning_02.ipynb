## Cleaning pitStops.csv

This section performs data cleaning and preprocessing for the pitStops dataset. The file contains Formula 1 pit stop records including race ID, driver ID, lap number, stop number, stop time, duration, and milliseconds.

```python
pit.head()
```

```python
pit.info()
```

```python
pit.isnull().sum()
```

```python
pit.duplicated().sum()
```

### Cleaning Process

The dataset contains no missing values and no duplicate records. Additional preprocessing is applied to standardize column names and convert data types for consistency.

```python
# Clean column names
pit.columns = pit.columns.str.strip()

# Remove duplicate rows
pit.drop_duplicates(inplace=True)

# Handle missing values
# Text columns
for col in pit.select_dtypes(include='object').columns:
    pit[col].fillna("Unknown", inplace=True)

# Numeric columns
for col in pit.select_dtypes(include=['int64', 'float64']).columns:
    pit[col].fillna(pit[col].median(), inplace=True)

# Convert time column
pit['time'] = pd.to_datetime(pit['time'], format='%H:%M:%S', errors='coerce').dt.time

# Convert duration column
pit['duration'] = pd.to_numeric(pit['duration'], errors='coerce')

# Convert integer columns
pit['raceId'] = pit['raceId'].astype(int)
pit['driverId'] = pit['driverId'].astype(int)
pit['stop'] = pit['stop'].astype(int)
pit['lap'] = pit['lap'].astype(int)
pit['milliseconds'] = pit['milliseconds'].astype(int)

pit.head()
```

```python
pit.info()
pit.describe()
```

### Result

The pitStops dataset has been successfully cleaned and standardized. It is now ready for merging, visualization, and further analysis.

## Cleaning qualifying.csv

This section performs data cleaning and preprocessing for the qualifying dataset. The file contains Formula 1 qualifying session records including qualifying positions, drivers, constructors, and lap times (Q1, Q2, Q3).

```python
qual.head()
```

```python
qual.info()
```

```python
qual.isnull().sum()
```

```python
qual.duplicated().sum()
```

### Cleaning Process

The dataset is checked for missing values, duplicate records, inconsistent column names, and incorrect data types. Missing qualifying times are expected for drivers who did not advance to later sessions.

```python
# Clean column names
qual.columns = qual.columns.str.strip()

# Remove duplicate rows
qual.drop_duplicates(inplace=True)

# Fill missing qualifying times
qual['q1'] = qual['q1'].astype(str).str.strip().fillna("No Time")
qual['q2'] = qual['q2'].astype(str).str.strip().fillna("No Time")
qual['q3'] = qual['q3'].astype(str).str.strip().fillna("No Time")

# Convert numeric columns
qual['qualifyId'] = qual['qualifyId'].astype(int)
qual['raceId'] = qual['raceId'].astype(int)
qual['driverId'] = qual['driverId'].astype(int)
qual['constructorId'] = qual['constructorId'].astype(int)
qual['number'] = qual['number'].astype(int)
qual['position'] = qual['position'].astype(int)

qual.head()
```

```python
qual.info()
qual.describe()
```

### Result

The qualifying dataset has been successfully cleaned and standardized. Missing lap times were handled appropriately, duplicate rows were removed, and column formats were corrected. The dataset is now ready for analysis and visualization.

## Cleaning races.csv

This section performs data cleaning and preprocessing for the races dataset. The file contains Formula 1 race event information including season year, round number, circuit, race name, race date, time, and related session schedules.

```python
races.head()
```

```python
races.info()
```

```python
races.isnull().sum()
```

```python
races.duplicated().sum()
```

### Cleaning Process

The dataset is checked for missing values, duplicate rows, incorrect data types, and inconsistent formatting. Date and time columns are converted into proper datetime format for easier analysis.

```python
# Clean column names
races.columns = races.columns.str.strip()

# Remove duplicate rows
races.drop_duplicates(inplace=True)

# Fill missing text values
races['name'] = races['name'].fillna("Unknown").str.strip()
races['url'] = races['url'].fillna("Unknown").str.strip()

# Convert date column
races['date'] = pd.to_datetime(races['date'].astype(str).str.strip(), errors='coerce')

# Handle all null values in time column
races['time'] = races['time'].fillna("Unknown").astype(str).str.strip()

# Convert numeric columns
races['raceId'] = races['raceId'].astype(int)
races['year'] = races['year'].astype(int)
races['round'] = races['round'].astype(int)
races['circuitId'] = races['circuitId'].astype(int)

races.head()
```

```python
races.info()
races.describe(include='all')
```

### Result

The races dataset has been successfully cleaned and standardized. Missing values were handled, duplicate rows removed, date fields corrected, and data formats improved. The dataset is now ready for analysis and visualization.

## Cleaning results.csv

This section performs data cleaning and preprocessing for the results dataset. The file contains Formula 1 race result records including driver standings, finishing positions, lap data, points earned, and race status.

```python
results.head()
```

```python
results.info()
```

```python
results.isnull().sum()
```

```python
results.duplicated().sum()
```

### Cleaning Process

The dataset is checked for missing values, duplicate rows, inconsistent formatting, and incorrect data types. Numeric fields are standardized and missing values are handled for smoother analysis.

```python
# Clean column names
results.columns = results.columns.str.strip()

# Remove duplicate rows
results.drop_duplicates(inplace=True)

# Handle missing values properly

# Text columns
results['time'].fillna("Unknown", inplace=True)
results['fastestLapTime'].fillna("Unknown", inplace=True)

# Numeric columns
results['number'].fillna(0, inplace=True)
results['position'].fillna(0, inplace=True)
results['milliseconds'].fillna(0, inplace=True)
results['fastestLap'].fillna(0, inplace=True)
results['rank'].fillna(0, inplace=True)
results['points'].fillna(0, inplace=True)

# Fastest lap speed (best to use median)
results['fastestLapSpeed'] = pd.to_numeric(results['fastestLapSpeed'], errors='coerce')
results['fastestLapSpeed'].fillna(results['fastestLapSpeed'].median(), inplace=True)

# Convert numeric columns
results['resultId'] = results['resultId'].astype(int)
results['raceId'] = results['raceId'].astype(int)
results['driverId'] = results['driverId'].astype(int)
results['constructorId'] = results['constructorId'].astype(int)
results['number'] = results['number'].astype(int)
results['grid'] = results['grid'].astype(int)
results['position'] = results['position'].astype(int)
results['positionOrder'] = results['positionOrder'].astype(int)
results['laps'] = results['laps'].astype(int)
results['milliseconds'] = results['milliseconds'].astype(int)
results['fastestLap'] = results['fastestLap'].astype(int)
results['rank'] = results['rank'].astype(int)
results['statusId'] = results['statusId'].astype(int)
results['points'] = results['points'].astype(float)

results.head()
```

```python
results.info()
results.describe(include='all')
```

### Result

The results dataset has been successfully cleaned and standardized. Duplicate records were removed, missing values handled, and numeric columns corrected. The dataset is now ready for analysis and visualization.

## Cleaning status.csv

This section performs data cleaning and preprocessing for the status dataset. The file contains Formula 1 race result status descriptions such as finished, accident, engine failure, disqualified, and other race outcomes.

```python
status.head()
```

```python
status.info()
```

```python
status.isnull().sum()
```

```python
status.duplicated().sum()
```

### Cleaning Process

The dataset is checked for missing values, duplicate rows, inconsistent formatting, and incorrect data types. Text values are standardized to improve readability and consistency.

```python
# Standardize column names
status.columns = status.columns.str.strip()

# Remove duplicate rows
status.drop_duplicates(inplace=True)

# Fill missing values
status.fillna("Unknown", inplace=True)

# Standardize text formatting
status['status'] = status['status'].str.strip().str.title()

# Convert numeric column
status['statusId'] = status['statusId'].astype(int)

status.head()
```
